### Imports


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

### Introducci贸n de datos

```python
df = pd.read_csv('RegresionLineal.csv')
```

#### Visualizaci贸n de datos
<span style="color:red">se debe llenar los espacios que corresponden a x_label y y_label para que el resto del c贸digo corra. Se puede dar formato a las unidades de las coordenadas con comandos de LaTeX entre signos de dolar</span>

```python
x = np.array(df['x']) #Convertir datos a arrays de Numpy (NECESARIO)
y = np.array(df['y'])

x_label = input("Etiqueta eje x:") # se pueden formatear unidades con comandos de LaTeX ej: $\frac{mol}{L}$
y_label = input("Etiqueta eje y:")

plt.figure(figsize=(5, 3), dpi=300)

plt.xlabel(x_label)
plt.ylabel(y_label)

plt.xlim(0,0.00025) #ajustar la escala en las abscisas
#plt.ylim(inicio, fin) ajuste en ordenadas de ser necesario

plt.scatter(x, y)

plt.show()  #
```

#### Tratamiento de datos
Se debe reacomodar los datos en un array de la dimensi贸n que especifica el paquete

```python
x_fit = x.reshape(-1, 1)
y_fit = y.reshape(-1, 1)
```

#### Llamando la funci贸n de regresi贸n con los par谩metros ajustados

```python
regresion = LinearRegression()
regresion.fit(x_fit, y_fit)
```

#### Elementos de regresi贸n

```python
R_cuadrado = regresion.score(x_fit,y_fit) #Coeficiente de correlaci贸n
pendiente = regresion.coef_#pendiente
intercepto = regresion.intercept_ #intercepto (b)
```

#### Obtenci贸n de tabla

```python
data = {r'Pendiente': [pendiente], r'$R^2$': [R_cuadrado], "Intercepto": [intercepto]}
tabla = pd.DataFrame(data)

tabla
```

#### visualizaci贸n de datos contra ajuste lineal

```python
#Extrapolaci贸n para l铆nea de mejor ajuste

y_fit_pred = regresion.predict(x_fit)
print(y_fit_pred)
```

```python
plt.figure(figsize=(5, 3), dpi=300)

plt.xlabel(x_label)
plt.ylabel(y_label)

plt.scatter(x, y)
plt.plot(x_fit, y_fit_pred, "pink")
plt.xlim(0,0.00025)

plt.show()  #
```
